```python
import pandas as pd

# Load the downloaded CSV (adjust filename/path if needed)
usd_inr_df = pd.read_csv('USD_INR Historical Data_AprMay2025.csv')

# Keep only Date and Price (USDINR)
usd_inr_df = usd_inr_df[['Date', 'Price']].rename(columns={'Price': 'USDINR'})

# Convert date to consistent format
usd_inr_df['Date'] = pd.to_datetime(usd_inr_df['Date'], dayfirst=True).dt.strftime('%d-%m-%Y')
lbma_df['Date'] = pd.to_datetime(lbma_df['Date'], dayfirst=True).dt.strftime('%d-%m-%Y')
ibja_df['Date'] = pd.to_datetime(ibja_df['Date'], dayfirst=True).dt.strftime('%d-%m-%Y')

usd_inr_df.head()
```

```python
# Merge LBMA and IBJA on Date
merged_df = pd.merge(lbma_df, ibja_df, on='Date', how='inner')

# Merge with USD-INR rates
merged_df = pd.merge(merged_df, usd_inr_df, on='Date', how='inner')

# # Display final merged DataFrame
merged_df.head()
```

```python
merged_df.info()
```

```python
# Define the conversion factor
oz_to_10g = 10 / 31.1035
#including import duty,gst and others
adjustment_factor = 1.051

# Apply the conversion
merged_df['IBMA_INR_10g_AM'] = merged_df['AM Price (USD/oz)'] * merged_df['USDINR'] * oz_to_10g * adjustment_factor
merged_df['IBMA_INR_10g_PM'] = merged_df['PM Price (USD/oz)'] * merged_df['USDINR'] * oz_to_10g * adjustment_factor

# Optional: round for cleaner display
merged_df['IBMA_INR_10g_AM'] = merged_df['IBMA_INR_10g_AM'].round(2)
merged_df['IBMA_INR_10g_PM'] = merged_df['IBMA_INR_10g_PM'].round(2)

merged_df.head(10)
```
